```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definition of vectors

n = 800

x = np.linspace(0,n,n)
y = np.linspace(0,n,n)
```

```python
x[0] = 1
y[0] = 1
h = 0.1

for i in range(0, n-1):
    x[i+1] = h + x[i]*x[i]*y[i]*h + (1 - 3.5*h)*x[i]
    y[i+1] = -h*x[i]*x[i]*y[i] + 2.5*h*x[i] + y[i]
```

```python
z = np.arange(0,n,1)

plt.plot(z, x[z], color='blue')
plt.plot(z, y[z], color='red')

plt.xlabel('t', fontsize='20')
plt.ylabel('x(t)', fontsize='20')
plt.tick_params(labelsize=15)
```

```python
plt.scatter(x, y, s=10, facecolors='none', edgecolors='red')
plt.ylabel(r'$y$', fontsize='20')
plt.xlabel(r'$x$', fontsize='20')
plt.ylim(0,5)
plt.xlim(0,3.5)
plt.tick_params(labelsize=15)
plt.title('Phase plane')
plt.scatter(1,2.5,s=20, facecolors='black', edgecolors='black', label='(1,2.5)')
plt.legend(loc='upper right')
plt.savefig('/Users/κωτς/Desktop/phaseplane.pdf', transparent=True, bbox_inches='tight')
```
