```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
a = 20

X = [0]
Y = [50]

t = [0]
```

```python
tend = 200
V = 10
Ca = 1
Cb = 7
Cc = 1
Na = Ca*V
Nb = Cb*V
Nc = Cc*V
k1p = 0.5
k1n = 0.25
k2p = 1
k2n = 0.25
k3p = 1
k3n = 0.25
```

```python
while t[-1] < tend:
    
    props = [ k1p*Ca*V , k1n*X[-1], k2p*Cb*X[-1], k2n*Cc*Y[-1]  ,(k3p*X[-1]*(X[-1]-1)*Y[-1])/(V**2), (k3n*X[-1]*(X[-1]-1)*(X[-1]-2))/(V**2) ]
    
    prop_sum = sum(props)
    
    if prop_sum == 0:
        break
    
    tau = np.random.exponential(scale=1/prop_sum)
    
    t.append(t[-1] + tau)
    
    rand = random.uniform(0,1)
    
    if rand*prop_sum <= props[0]:
        X.append(X[-1] + 1)
        Y.append(Y[-1])
        
    elif rand*prop_sum > props[0] and rand*prop_sum <= props[0]+props[1]:
        X.append(X[-1] - 1)
        Y.append(Y[-1])   
        
    elif rand*prop_sum > props[0]+props[1] and rand*prop_sum <= props[0]+props[1]+props[2]:
        X.append(X[-1] - 1)
        Y.append(Y[-1] + 1)  
        
    elif rand*prop_sum > props[0]+props[1]+props[2] and rand*prop_sum <= props[0]+props[1]+props[2]+props[3]:
        X.append(X[-1] + 1)
        Y.append(Y[-1] - 1) 
        
    elif rand*prop_sum > props[0]+props[1]+props[2]+props[3] and rand*prop_sum <= props[0]+props[1]+props[2]+props[3]+props[4]:
        X.append(X[-1] + 1)
        Y.append(Y[-1] - 1) 
        
    elif rand*prop_sum > props[0]+props[1]+props[2]+props[3]+props[4] and rand*prop_sum <= props[0]+props[1]+props[2]+props[3]+props[4]+props[5]:
        X.append(X[-1] - 1)
        Y.append(Y[-1] + 1) 
```

```python
X_plot, = plt.plot(t, X, label="X", color='blue')
#Y_plot, = plt.plot(t, Y, label="Y", color='red')

#plt.legend(handles=[X_plot, Y_plot])
plt.xlabel('t', fontsize='20')
plt.ylabel('X(t)', fontsize='20')
#plt.xlim(0,800)
#plt.ylim(0,10)
plt.tick_params(labelsize=15)
```
